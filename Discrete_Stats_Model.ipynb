# Discrete Statistical Distribution

### Binomial Plotting

```python
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import binom
import pandas as pd
```

```python
n = 10
p = 0.8

x = list(range(n + 1))
pmf = binom.pmf(x, n, p)
df = pd.DataFrame({
    "# of Successes": x,
    "Probability": pmf
})

plt.figure(figsize=(8, 5))
sns.barplot(data=df, x="# of Successes", y="Probability", color="deepskyblue")
plt.title("Binomial Distribution (n=10, p=0.8)")
plt.xlabel("# of Successes")
plt.ylabel("Probability")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

### Geometric Plotting

```python
from scipy.stats import geom

p = 0.8
x = list(range(10))  # 0 to 9 failures
pmf = geom.pmf([i+1 for i in x], p)  # geom in scipy counts trials (starts at 1), so shift x by +
df = pd.DataFrame({
    "Failures Before First Success": x,
    "Probability": pmf
})

plt.figure(figsize=(8, 5))
sns.barplot(data=df, x="Failures Before First Success", y="Probability", color="deepskyblue")
plt.title("Geometric Distribution (p = 0.8)\nFailures Before First Success")
plt.xlabel("Failures Before First Success")
plt.ylabel("Probability")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
